```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the quality score file
file_name = "Book3.txt"  # Change if your file has a different name

# Read the file assuming tab or space separation
df = pd.read_csv(file_name, sep="\t", header=None, names=["Quality Score", "Frequency"])

# Convert to numeric (handle any unexpected non-numeric values)
df["Quality Score"] = pd.to_numeric(df["Quality Score"], errors="coerce")
df["Frequency"] = pd.to_numeric(df["Frequency"], errors="coerce")

# Remove rows with missing values
df = df.dropna()

# Plot the quality score distribution
plt.figure(figsize=(10, 5))  # Set figure size

# Use a line plot instead of a bar plot if there are many data points
sns.barplot(x=df["Quality Score"], y=df["Frequency"], color="blue")

# Set labels and title
plt.xlabel("Quality Score (QUAL)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Quality Score Distribution in AD VCF Variants", fontsize=14)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Add a grid for better readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the quality score file
file_name = "Book3.txt"  # Change if your file has a different name

# Read the file assuming tab or space separation
df = pd.read_csv(file_name, sep="\t", header=None,
                 names=["Quality Score", "Frequency"])

# Convert to numeric (handle any unexpected non-numeric values)
df["Quality Score"] = pd.to_numeric(df["Quality Score"], errors="coerce")
df["Frequency"] = pd.to_numeric(df["Frequency"], errors="coerce")

# Remove rows with missing values
df = df.dropna()

# Plot the quality score distribution
plt.figure(figsize=(10, 5))  # Set figure size

# Use a bar plot
sns.barplot(x=df["Quality Score"], y=df["Frequency"], color="blue")

# Set labels and title
plt.xlabel("Quality Score (QUAL)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Quality Score Distribution in AD VCF Variants", fontsize=14)

# Rotate x-axis labels for better visibility and adjust tick frequency
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and align right
plt.xticks(range(0, len(df["Quality Score"]), len(
    df["Quality Score"]) // 10))  # Adjust x-axis ticks

# Add a grid for better readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()  # Ensure the layout fits well
plt.savefig("quality_score_distribution.png") #save as png file
plt.show()
```
