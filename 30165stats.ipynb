```python
import matplotlib.pyplot as plt

# Read the quality scores and their frequencies from the file
quality_scores = []
frequencies = []

with open('quality_scores.txt', 'r') as file:
    for line in file:
        # Split each line into frequency and quality score
        freq, score = line.split()
        quality_scores.append(float(score))
        frequencies.append(int(freq))

# Plotting the frequency plot
plt.figure(figsize=(10, 6))
plt.bar(quality_scores, frequencies, color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Frequency of Quality Scores', fontsize=16)
plt.xlabel('Quality Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
plt.savefig("quality_score_distribution.png")  # save as png file
```
